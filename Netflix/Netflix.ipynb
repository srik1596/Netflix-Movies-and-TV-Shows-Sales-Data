```python
import pandas as pd

df = pd.read_csv('netflix_titles.csv')
df.head()
```

```python
df.shape
```

```python
#1.Identify and handle missing values
df.isnull().sum()
```

```python
#Handled missing values
df['director'] = df['director'].fillna('Not available')
df['cast'] = df['cast'].fillna('Not available')
df['country'] = df['country'].fillna('Not available')
df['date_added'] = df['date_added'].fillna('Not available')
df['rating'] = df['rating'].fillna(df['rating'].mode()[0]) #most common rating
df['duration'] = df['duration'].fillna('unknown')
```

```python
df.isnull().sum()
```

```python
#2.Remove duplicate rows
df.drop_duplicates()
```

```python
df.columns
```

```python
#3.Standardize text values
text_columns = ['type', 'title', 'director', 'cast', 'country', 'listed_in', 'description']
for col in text_columns:
    df[col] = df[col].str.strip().str.title()
```

```python
#4.Convert date formats to a consistent type
df['date_added'] = pd.to_datetime(df['date_added'], errors='coerce', dayfirst=True)
```

```python
#5.Rename column headers to be clean and uniform
df.columns = df.columns.str.strip().str.title()
```
